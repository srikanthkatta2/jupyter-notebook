```python
import boto3

# Create an S3 client

s3 = boto3.client("s3")
```

```python
bucket_name = "mynlpmru3"

s3.create_bucket(

    Bucket=bucket_name,

    CreateBucketConfiguration={"LocationConstraint": "eu-north-1"}

)

print(f"Bucket '{bucket_name}' created successfully!")
```
